```python
# Imports
import pandas as pd
from datetime import datetime
from math import sin, cos, sqrt, atan2, radians
```

```python
jan_2021 = "data/JC-202101-citibike-tripdata.csv"
feb_2021 = "data/JC-202102-citibike-tripdata.csv"
mar_2021 = "data/JC-202103-citibike-tripdata.csv"
apr_2021 = "data/JC-202104-citibike-tripdata.csv"
may_2021 = "data/JC-202105-citibike-tripdata.csv"
jun_2021 = "data/JC-202106-citibike-tripdata.csv"
jul_2021 = "data/JC-202107-citibike-tripdata.csv"
aug_2021 = "data/JC-202108-citibike-tripdata.csv"
sep_2021 = "data/JC-202109-citibike-tripdata.csv"
oct_2021 = "data/JC-202110-citibike-tripdata.csv"
nov_2021 = "data/JC-202111-citibike-tripdata.csv"
dec_2021 = "data/JC-202112-citibike-tripdata.csv"
```

```python
# Read Feb-Dec data in the same format
df_feb_2021 = pd.read_csv(feb_2021, encoding="ISO-8859-1")
df_mar_2021 = pd.read_csv(mar_2021, encoding="ISO-8859-1")
df_apr_2021 = pd.read_csv(apr_2021, encoding="ISO-8859-1")
df_may_2021 = pd.read_csv(may_2021, encoding="ISO-8859-1")
df_jun_2021 = pd.read_csv(jun_2021, encoding="ISO-8859-1")
df_jul_2021 = pd.read_csv(jul_2021, encoding="ISO-8859-1")
df_aug_2021 = pd.read_csv(aug_2021, encoding="ISO-8859-1")
df_sep_2021 = pd.read_csv(sep_2021, encoding="ISO-8859-1")
df_oct_2021 = pd.read_csv(oct_2021, encoding="ISO-8859-1")
df_nov_2021 = pd.read_csv(nov_2021, encoding="ISO-8859-1")
df_dec_2021 = pd.read_csv(dec_2021, encoding="ISO-8859-1")
```

```python
# Combining all months csv files
citibike_2021_df = df_feb_2021.append(df_mar_2021, sort=False)
citibike_2021_df = citibike_2021_df.append(df_apr_2021, sort=False)
citibike_2021_df = citibike_2021_df.append(df_may_2021, sort=False)
citibike_2021_df = citibike_2021_df.append(df_jun_2021, sort=False)
citibike_2021_df = citibike_2021_df.append(df_jul_2021, sort=False)
citibike_2021_df = citibike_2021_df.append(df_aug_2021, sort=False)
citibike_2021_df = citibike_2021_df.append(df_sep_2021, sort=False)
citibike_2021_df = citibike_2021_df.append(df_oct_2021, sort=False)
citibike_2021_df = citibike_2021_df.append(df_nov_2021, sort=False)
citibike_2021_df = citibike_2021_df.append(df_dec_2021, sort=False)
```

```python
citibike_2021_df.dtypes
```

```python
citibike_2021_df.started_at = citibike_2021_df.started_at.str.replace(' ', '')
citibike_2021_df.started_at = citibike_2021_df.started_at.str.replace('-', '')
citibike_2021_df.started_at = citibike_2021_df.started_at.str.replace(':', '')
citibike_2021_df.ended_at = citibike_2021_df.ended_at.str.replace(' ', '')
citibike_2021_df.ended_at = citibike_2021_df.ended_at.str.replace('-', '')
citibike_2021_df.ended_at = citibike_2021_df.ended_at.str.replace(':', '')
```

```python
citibike_2021_df.started_at = pd.to_datetime(citibike_2021_df.started_at, format='%Y%m%d%H%M%S')
citibike_2021_df['started_date'] = citibike_2021_df.started_at.dt.date
citibike_2021_df.ended_at = pd.to_datetime(citibike_2021_df.ended_at, format='%Y%m%d%H%M%S')
citibike_2021_df['ended_date'] = citibike_2021_df.ended_at.dt.date

citibike_2021_df.head(2)
```

```python
citibike_2021_df.dtypes
```

```python
citibike_2021_df['tripduration'] = (citibike_2021_df.ended_at - citibike_2021_df.started_at)
citibike_2021_df['tripduration'] = round(citibike_2021_df['tripduration'].dt.seconds/60,2)
citibike_2021_df=citibike_2021_df[['started_at','ended_at','start_station_name','start_station_id','end_station_name','end_station_id','start_lat','start_lng','end_lat','end_lng','member_casual','tripduration','started_date','ended_date']]
citibike_2021_df.head(2)
```

```python
df_jan_2021 = pd.read_csv(jan_2021, encoding="ISO-8859-1")
```

```python
df_jan_2021 = df_jan_2021[['starttime','stoptime','start station name','start station id','end station name','end station id','start station latitude','start station longitude','end station latitude','end station longitude','usertype','tripduration']]
df_jan_2021.rename(columns={"starttime":"started_at",
                           "stoptime":"ended_at",
                           "start station name":"start_station_name",
                           "start station id":"start_station_id",
                           "end station name":"end_station_name",
                           "end station id":"end_station_id",
                           "start station latitude":"start_lat",
                           "start station longitude":"start_lng",
                           "end station latitude":"end_lat",
                           "end station longitude":"end_lng",
                           "usertype":"member_casual"
                        }, inplace=True)
df_jan_2021.tripduration = round(df_jan_2021.tripduration/60,2)
df_jan_2021.head(2)
```

```python
df_jan_2021.started_at = pd.to_datetime(df_jan_2021['started_at'], infer_datetime_format=True, format='%Y%m%d')
df_jan_2021.ended_at = pd.to_datetime(df_jan_2021['ended_at'], infer_datetime_format=True, format='%Y%m%d')
df_jan_2021['started_date'] = df_jan_2021.started_at.dt.date
df_jan_2021['ended_date'] = df_jan_2021.ended_at.dt.date
df_jan_2021.head(2)
```

```python
df_jan_2021.member_casual = df_jan_2021.member_casual.str.replace('Subscriber', 'member')
df_jan_2021.member_casual = df_jan_2021.member_casual.str.replace('Customer', 'casual')
```

```python
citibike_2021_df = citibike_2021_df.append(df_jan_2021, sort=False)
```

```python
# drop na values
citibike_2021_df.dropna(subset=["end_lat", "end_lng", "end_station_name"], inplace = True)
```

```python
# break out Year/Month
citibike_2021_df['trip_year'] = pd.to_datetime(citibike_2021_df.started_date).dt.year
citibike_2021_df['trip_month'] = pd.to_datetime(citibike_2021_df.started_date).dt.month
```

```python
citibike_2021_df.trip_month.replace({1: "January",
                                2: "February",
                                3: "March",
                                4: "April",
                                5: "May",
                                6: "June",
                                7: "July",
                                8: "August",
                                9: "September",
                                10: "October",
                                11: "November",
                                12: "December"}, inplace=True)
```

```python
citibike_2021_df.start_lat = round(citibike_2021_df.start_lat,3)
citibike_2021_df.start_lng = round(citibike_2021_df.start_lng,3)
citibike_2021_df.end_lat = round(citibike_2021_df.end_lat,3)
citibike_2021_df.end_lng = round(citibike_2021_df.end_lng,3)
```

```python
citibike_2021_df.head(2)
```

```python
# Convert coordiantes to earth distance
lat_started = pd.Series([radians(lat) for lat in citibike_2021_df["start_lat"]])
lon_started = pd.Series([radians(lon) for lon in citibike_2021_df["start_lng"]])
lat_ended = pd.Series([radians(lat) for lat in citibike_2021_df["end_lat"]])
lon_ended = pd.Series([radians(lon) for lon in citibike_2021_df["end_lng"]])

distance_lat = lat_ended - lat_started
distance_lon = lon_ended - lon_started
```

```python
# Approximate radius of Earth in km
R = 6373.0

trip_distance = []

for i in range(0, len(lat_started)):
    
    a = sin(distance_lat[i] / 2)**2 + cos(lat_started[i]) * cos(lat_ended[i]) * sin(distance_lon[i] / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Calculate and convert km to miles
    miles = (R * c) * .6214
    
    trip_distance.append(miles)
```

```python
citibike_2021_df["trip_distance(mi)"] = trip_distance
```

```python
citibike_2021_df['Start_End_Stations'] = citibike_2021_df[['start_station_name', 'end_station_name']].agg(' to '.join, axis=1)
```
